```python
import pandas as pd

def calculate_latencies(df):
    df["network_latency"] = 0.720
    df["save_latency"] = (df["image_save_at"] - df["server_receive_at"]) * 1000
    df["preprocess_latency"] = (df["image_preprocessed_at"] - df["image_save_at"]) * 1000
    df["inference_latency"] = (df["image_predicted_at"] - df["image_preprocessed_at"]) * 1000
    df["total_latency"] = df["network_latency"] + df["save_latency"] + df["preprocess_latency"] + df["inference_latency"]
    df["rtt"] = (df["client_receive_at"] - df["client_send_at"]) * 1000
    df["rtt - total_latency"] = df["rtt"] - df["total_latency"]

    return df[["network_latency", "save_latency", "preprocess_latency", "inference_latency", "total_latency", "rtt", "rtt - total_latency"]]

mobilenet_v3_small = pd.read_csv("/Users/neeleshkarthikeyan/d2i/ckn-edge-optimization/micro_benchmarking/results/mobilenet_v3_small.csv")
mobilenet_v3_small = calculate_latencies(mobilenet_v3_small)
mobilenet_v3_small
```
